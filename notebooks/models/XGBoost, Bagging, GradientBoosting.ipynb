Nous avons téléchargé nos jeux d'entraînement et de test sous forme de csv afin de les récupérer directement sans avoir à relancer tout le preprocessing. Les paramètres choisies sont l'année 2010, 20 notes possibles pour la difficulté.

```python
import pandas as pd
import numpy as np
from sklearn.metrics import classification_report
X_train = pd.read_csv("./../../data/X_train.csv",index_col = 0)
X_test = pd.read_csv("./../../data/X_test.csv",index_col = 0)
y_train = pd.read_csv("./../../data/y_train.csv",index_col= 0)
y_test = pd.read_csv("./../../data/y_test.csv",index_col = 0)
```

## Gradient Boosting

```python
from sklearn.ensemble import GradientBoostingClassifier
gb = GradientBoostingClassifier(n_estimators = 100,learning_rate = 0.1,max_depth = 10,random_state=0)
gb.fit(X_train,y_train)
print(classification_report(y_test,gb.predict(X_test)))
```

## Bagging

```python
from sklearn.ensemble import BaggingClassifier
bc = BaggingClassifier(n_estimators=100,oob_score=True)
bc.fit(X_train,y_train)
print(classification_report(y_test,bc.predict(X_test)))
```

## XGBoost

```python
import xgboost as xgb
train = xgb.DMatrix(data=X_train,label=y_train)
test_xgb=xgb.DMatrix(data=X_test,label=y_test)
params = {'booster' : 'gbtree','learning_rate' : 0.01, 'binary' : 'logistic'}
xgb = xgb.train(params,num_boost_round=700,early_stopping_rounds=15,evals=[(train, 'train'), (test_xgb, 'eval')],dtrain=train)
preds=xgb.predict(test_xgb)
xgbpreds = pd.Series([1 if p>0.5 else 0 for p in preds])
print(classification_report(y_test,xgbpreds))
```
